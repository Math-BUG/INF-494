## Datasets

Os datasets abaixo foram escolhidos para expor situações diferentes do DBSCAN:

- `dense_blobs_2d`: blobs densos 2D, bons para mostrar muitos core points.
- `heterogeneous_blobs_2d`: regiões com densidades diferentes.

```python
# ============================================================
# 3. Configuracao dos datasets
# ============================================================

# A implementacao Numba abaixo usa vizinhanca por forca bruta: custo O(n^2).
# Use "quick" para depurar no Colab e "presentation" para resultados mais
# proximos do cenario discutido pelo professor, com mais trabalho para a GPU.
RUN_MODE = "quick"  # opcoes: "quick" ou "presentation"

if RUN_MODE == "quick":
    N_SAMPLES_2D = 1800
    N_SAMPLES_32D = 900
elif RUN_MODE == "presentation":
    N_SAMPLES_2D = 6000
    N_SAMPLES_32D = 1800
else:
    raise ValueError("RUN_MODE deve ser 'quick' ou 'presentation'.")

MIN_SAMPLES_DEFAULT = 8

DATASET_CONFIGS = [
    ("dense_blobs_2d", N_SAMPLES_2D, "Blobs densos 2D", 0.90),
    ("heterogeneous_blobs_2d", N_SAMPLES_2D, "Blobs 2D com densidades diferentes", 0.90),
    ("dense_blobs_noise_2d", N_SAMPLES_2D, "Blobs densos 2D com ruido/outliers", 0.90),
    ("moons_2d", N_SAMPLES_2D, "Duas luas 2D", 0.88),
    ("blobs_32d", N_SAMPLES_32D, "Blobs com 32 dimensoes para testar 8/4 bits", 0.90),
]

print("RUN_MODE:", RUN_MODE)
print("N_SAMPLES_2D:", N_SAMPLES_2D)
print("N_SAMPLES_32D:", N_SAMPLES_32D)
```

```python
# ============================================================
# 4. Funções para gerar e normalizar datasets
# ============================================================

def normalize_minmax(X):
    """Normaliza cada atributo para [0, 1]."""
    X = np.asarray(X, dtype=np.float32)
    mn = X.min(axis=0)
    mx = X.max(axis=0)
    denom = mx - mn
    denom[denom == 0.0] = 1.0
    return ((X - mn) / denom).astype(np.float32)


def estimate_eps(X, min_samples=8, quantile=0.90, sample_size=5000, seed=SEED):
    """Estima eps pela distância até o k-ésimo vizinho."""
    rng = np.random.default_rng(seed)
    X = np.asarray(X, dtype=np.float32)

    if len(X) > sample_size:
        idx = rng.choice(len(X), size=sample_size, replace=False)
        Xs = X[idx]
    else:
        Xs = X

    nn = NearestNeighbors(n_neighbors=min_samples)
    nn.fit(Xs)
    dists, _ = nn.kneighbors(Xs)
    kth_dist = dists[:, -1]
    return float(np.quantile(kth_dist, quantile))


def make_dataset(name, n_samples, seed=SEED):
    """
    Cria datasets sintéticos controlados.

    Datasets densos são importantes porque geram muitos core points. Quando há muitos
    cores, a conexão entre eles fica cara por causa da sobreposição de vizinhanças.
    """
    rng = np.random.default_rng(seed)

    if name == "dense_blobs_2d":
        X, y = make_blobs(
            n_samples=n_samples,
            centers=[(-3, -3), (-3, 3), (3, -3), (3, 3)],
            cluster_std=0.18,
            random_state=seed,
        )

    elif name == "heterogeneous_blobs_2d":
        n1 = int(0.40 * n_samples)
        n2 = int(0.35 * n_samples)
        n3 = n_samples - n1 - n2
        X1, _ = make_blobs(n_samples=n1, centers=[(-4, 0)], cluster_std=0.08, random_state=seed + 1)
        X2, _ = make_blobs(n_samples=n2, centers=[(0, 0)], cluster_std=0.22, random_state=seed + 2)
        X3, _ = make_blobs(n_samples=n3, centers=[(4, 0)], cluster_std=0.55, random_state=seed + 3)
        X = np.vstack([X1, X2, X3])
        y = np.concatenate([
            np.zeros(n1, dtype=np.int32),
            np.ones(n2, dtype=np.int32),
            np.full(n3, 2, dtype=np.int32),
        ])

    elif name == "dense_blobs_noise_2d":
        n_noise = int(0.20 * n_samples)
        n_blob = n_samples - n_noise
        X_blob, y_blob = make_blobs(
            n_samples=n_blob,
            centers=[(-3, -3), (-3, 3), (3, -3), (3, 3)],
            cluster_std=0.20,
            random_state=seed,
        )
        noise = rng.uniform(low=-5.5, high=5.5, size=(n_noise, 2)).astype(np.float32)
        X = np.vstack([X_blob, noise])
        y = np.concatenate([y_blob.astype(np.int32), np.full(n_noise, -1, dtype=np.int32)])

    elif name == "moons_2d":
        X, y = make_moons(n_samples=n_samples, noise=0.045, random_state=seed)

    elif name == "blobs_32d":
        X, y = make_blobs(
            n_samples=n_samples,
            centers=6,
            n_features=32,
            cluster_std=0.45,
            random_state=seed,
        )

    else:
        raise ValueError(f"Dataset desconhecido: {name}")

    return normalize_minmax(X), np.asarray(y, dtype=np.int32)
```

```python
# ============================================================
# 5. Criar todos os datasets no dicionário datasets
# ============================================================

datasets = {}

for name, n_samples, description, eps_quantile in DATASET_CONFIGS:
    X_tmp, y_tmp = make_dataset(name=name, n_samples=n_samples, seed=SEED)
    eps_tmp = estimate_eps(
        X_tmp,
        min_samples=MIN_SAMPLES_DEFAULT,
        quantile=eps_quantile,
        seed=SEED,
    )

    datasets[name] = {
        "X": np.ascontiguousarray(X_tmp.astype(np.float32)),
        "y_true": y_tmp,
        "eps": eps_tmp,
        "min_samples": MIN_SAMPLES_DEFAULT,
        "description": description,
        "eps_quantile": eps_quantile,
    }

df_datasets = pd.DataFrame([
    {
        "dataset": name,
        "shape": datasets[name]["X"].shape,
        "eps": datasets[name]["eps"],
        "min_samples": datasets[name]["min_samples"],
        "descrição": datasets[name]["description"],
    }
    for name in datasets
])

display(df_datasets)
```

## Avaliação e baseline

O baseline é a referência para as métricas ARI e NMI. Se `cuML` estiver disponível, ele é usado. Caso contrário, o notebook usa `sklearn`.


```python
# ============================================================
# 7. Métricas, relabeling e visualização
# ============================================================

def count_clusters(labels):
    labels = np.asarray(labels)
    unique = set(labels.tolist())
    unique.discard(-1)
    return len(unique)


def noise_ratio(labels):
    labels = np.asarray(labels)
    return float(np.mean(labels == -1))


def relabel_consecutive(labels):
    """Converte labels arbitrários para 0, 1, 2, ... mantendo -1 como ruído."""
    labels = np.asarray(labels, dtype=np.int32)
    out = np.full(labels.shape, -1, dtype=np.int32)
    valid = [int(v) for v in np.unique(labels) if int(v) != -1]
    mapping = {old: new for new, old in enumerate(sorted(valid))}
    for old, new in mapping.items():
        out[labels == old] = new
    return out


def build_result_row(version, labels, elapsed, baseline_labels, info=None):
    info = {} if info is None else dict(info)
    row = {
        "versao": version,
        "tempo_s": float(elapsed),
        "clusters": count_clusters(labels),
        "ruido_%": 100.0 * noise_ratio(labels),
        "ARI_vs_baseline": adjusted_rand_score(baseline_labels, labels),
        "NMI_vs_baseline": normalized_mutual_info_score(baseline_labels, labels),
        "n_core_points": info.get("n_core_points", np.nan),
        "n_core_kept": info.get("n_core_kept", np.nan),
        "keep_ratio_real": info.get("keep_ratio_real", np.nan),
    }

    for key, value in info.items():
        if key not in row:
            row[key] = value
    return row


def plot_labels_2d(ax, X_plot, labels, title, max_points=6000, seed=SEED):
    if X_plot.shape[1] != 2:
        ax.axis("off")
        ax.set_title("Dataset não é 2D")
        return

    rng = np.random.default_rng(seed)
    n = X_plot.shape[0]
    if n > max_points:
        idx = rng.choice(n, size=max_points, replace=False)
        Xs = X_plot[idx]
        ls = np.asarray(labels)[idx]
    else:
        Xs = X_plot
        ls = np.asarray(labels)

    ax.scatter(Xs[:, 0], Xs[:, 1], c=ls, cmap="tab20", s=6, alpha=0.85, linewidths=0)
    ax.set_title(title)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
```

```python
# ============================================================
# 8. Função do baseline DBSCAN
# ============================================================

def _to_numpy(values):
    """Converte saída de cuML/cupy/cudf/sklearn para np.ndarray."""
    if values is None:
        return None
    if hasattr(values, "get"):
        return values.get()
    if hasattr(values, "to_numpy"):
        return values.to_numpy()
    if cp is not None:
        try:
            return cp.asnumpy(values)
        except Exception:
            pass
    return np.asarray(values)


def run_baseline_dbscan(X, eps, min_samples):
    X32 = np.ascontiguousarray(X.astype(np.float32))
    start = time.perf_counter()

    if HAS_CUML:
        X_gpu = cp.asarray(X32, dtype=cp.float32)
        model = CuMLDBSCAN(eps=float(eps), min_samples=int(min_samples))
        labels_raw = model.fit_predict(X_gpu)
        cp.cuda.Stream.null.synchronize()
        labels = _to_numpy(labels_raw).astype(np.int32)
        backend = "cuML"
    else:
        model = SklearnDBSCAN(eps=float(eps), min_samples=int(min_samples), n_jobs=-1)
        labels = model.fit_predict(X32).astype(np.int32)
        backend = "sklearn_cpu"

    elapsed = time.perf_counter() - start
    labels = relabel_consecutive(labels)

    core_indices = getattr(model, "core_sample_indices_", None)
    if core_indices is None:
        n_core = np.nan
    else:
        try:
            n_core = int(len(_to_numpy(core_indices)))
        except Exception:
            n_core = np.nan

    info = {
        "backend": backend,
        "n_core_points": n_core,
        "n_core_kept": n_core,
        "keep_ratio_real": 1.0 if not pd.isna(n_core) else np.nan,
        "quantizacao": "float32",
    }

    return labels, elapsed, backend, info
```

## DBSCAN Numba CUDA `float32` e drop de core points

A versão didática em GPU usa busca de vizinhos por força bruta. O drop mantém uma fração dos core points (`100%`, `75%`, `50%`, `25%`) escolhida de forma determinística por `seed`. Depois, pontos não rotulados são atribuídos ao core point mantido mais próximo dentro de `eps`.


```python
# ============================================================
# 9. Kernels comuns e versão float32
# ============================================================

TPB = 128


@cuda.jit
def count_core_float32_kernel(X, eps2, min_samples, core_flags, neighbor_counts):
    i = cuda.grid(1)
    n = X.shape[0]
    d = X.shape[1]

    if i >= n:
        return

    count = 0
    for j in range(n):
        dist2 = 0.0
        for k in range(d):
            diff = X[i, k] - X[j, k]
            dist2 += diff * diff
            if dist2 > eps2:
                break
        if dist2 <= eps2:
            count += 1
            if count >= min_samples:
                break

    neighbor_counts[i] = count
    if count >= min_samples:
        core_flags[i] = 1
    else:
        core_flags[i] = 0


@cuda.jit
def init_labels_kernel(core_flags, keep_mask, labels):
    i = cuda.grid(1)
    if i >= labels.shape[0]:
        return
    if core_flags[i] == 1 and keep_mask[i] == 1:
        labels[i] = i
    else:
        labels[i] = -1


@cuda.jit
def propagate_core_labels_float32_kernel(X, eps2, keep_mask, labels, changed):
    i = cuda.grid(1)
    n = X.shape[0]
    d = X.shape[1]

    if i >= n:
        return
    if keep_mask[i] == 0 or labels[i] < 0:
        return

    current_label = labels[i]
    best_label = current_label

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0.0
        for k in range(d):
            diff = X[i, k] - X[j, k]
            dist2 += diff * diff
            if dist2 > eps2:
                break

        if dist2 <= eps2:
            lj = labels[j]
            if lj >= 0 and lj < best_label:
                best_label = lj

    if best_label < current_label:
        labels[i] = best_label
        changed[0] = 1


@cuda.jit
def assign_points_to_nearest_core_float32_kernel(X, eps2, keep_mask, labels):
    i = cuda.grid(1)
    n = X.shape[0]
    d = X.shape[1]

    if i >= n:
        return
    if labels[i] >= 0:
        return

    best_label = -1
    best_dist2 = eps2
    found = 0

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0.0
        for k in range(d):
            diff = X[i, k] - X[j, k]
            dist2 += diff * diff
            if dist2 > eps2:
                break

        if dist2 <= eps2 and (found == 0 or dist2 < best_dist2):
            found = 1
            best_dist2 = dist2
            best_label = labels[j]

    labels[i] = best_label
```

```python
# ============================================================
# 10. Seleção determinística de cores e wrapper float32
# ============================================================

def make_core_keep_mask(core_flags, keep_ratio=1.0, seed=SEED):
    """
    Seleciona deterministamente uma fração dos core points.

    Em datasets densos, o número de core points pode ser muito grande. Manter só
    parte deles reduz a quantidade de conexões testadas entre cores, mas pode
    degradar a qualidade do agrupamento.
    """
    core_flags = np.asarray(core_flags, dtype=np.int32)
    keep_mask = np.zeros_like(core_flags, dtype=np.int32)
    core_idx = np.where(core_flags == 1)[0]

    if core_idx.size == 0:
        return keep_mask
    if keep_ratio >= 1.0:
        keep_mask[core_idx] = 1
        return keep_mask
    if keep_ratio <= 0.0:
        raise ValueError("keep_ratio deve ser maior que 0.")

    rng = np.random.default_rng(seed)
    n_keep = max(1, int(round(core_idx.size * float(keep_ratio))))
    selected = rng.permutation(core_idx)[:n_keep]
    keep_mask[selected] = 1
    return keep_mask


def dbscan_numba_float32(X, eps, min_samples, keep_ratio=1.0, max_iter=32, seed=SEED):
    if not cuda.is_available():
        raise RuntimeError("CUDA não disponível pelo Numba.")

    start = time.perf_counter()
    X32 = np.ascontiguousarray(X.astype(np.float32))
    n = X32.shape[0]
    blocks = max(1, (n + TPB - 1) // TPB)
    eps2 = np.float32(float(eps) * float(eps))

    d_X = cuda.to_device(X32)
    d_core_flags = cuda.device_array(n, dtype=np.int32)
    d_neighbor_counts = cuda.device_array(n, dtype=np.int32)

    count_core_float32_kernel[blocks, TPB](d_X, eps2, int(min_samples), d_core_flags, d_neighbor_counts)
    cuda.synchronize()

    core_flags = d_core_flags.copy_to_host()
    keep_mask = make_core_keep_mask(core_flags, keep_ratio=keep_ratio, seed=seed)

    d_keep_mask = cuda.to_device(keep_mask)
    d_labels = cuda.device_array(n, dtype=np.int32)
    init_labels_kernel[blocks, TPB](d_core_flags, d_keep_mask, d_labels)
    cuda.synchronize()

    iterations = 0
    for _ in range(int(max_iter)):
        iterations += 1
        d_changed = cuda.to_device(np.zeros(1, dtype=np.int32))
        propagate_core_labels_float32_kernel[blocks, TPB](d_X, eps2, d_keep_mask, d_labels, d_changed)
        cuda.synchronize()
        if d_changed.copy_to_host()[0] == 0:
            break

    assign_points_to_nearest_core_float32_kernel[blocks, TPB](d_X, eps2, d_keep_mask, d_labels)
    cuda.synchronize()

    labels = relabel_consecutive(d_labels.copy_to_host())
    elapsed = time.perf_counter() - start

    n_core = int(core_flags.sum())
    n_kept = int(keep_mask.sum())
    info = {
        "n_core_points": n_core,
        "n_core_kept": n_kept,
        "keep_ratio_real": float(n_kept / max(n_core, 1)),
        "iterations": iterations,
        "keep_ratio_target": float(keep_ratio),
        "quantizacao": "float32",
    }
    return labels, elapsed, info
```

## Versão `uint8`

Os dados são normalizados para `[0, 1]`, convertidos para inteiros de `0` a `255`, e o `eps` é multiplicado por `255`. Isso reduz leitura de memória por atributo de 4 bytes (`float32`) para 1 byte (`uint8`), com perda de precisão.


```python
# ============================================================
# 11. Quantização uint8 e kernels
# ============================================================

def quantize_uint8(X):
    Q = np.rint(np.clip(X, 0.0, 1.0) * 255.0).astype(np.uint8)
    return np.ascontiguousarray(Q)


@cuda.jit
def count_core_uint8_kernel(Q, eps2_q, min_samples, core_flags, neighbor_counts):
    i = cuda.grid(1)
    n = Q.shape[0]
    d = Q.shape[1]
    if i >= n:
        return

    count = 0
    for j in range(n):
        dist2 = 0
        for k in range(d):
            diff = int(Q[i, k]) - int(Q[j, k])
            dist2 += diff * diff
            if dist2 > eps2_q:
                break
        if dist2 <= eps2_q:
            count += 1
            if count >= min_samples:
                break

    neighbor_counts[i] = count
    if count >= min_samples:
        core_flags[i] = 1
    else:
        core_flags[i] = 0


@cuda.jit
def propagate_core_labels_uint8_kernel(Q, eps2_q, keep_mask, labels, changed):
    i = cuda.grid(1)
    n = Q.shape[0]
    d = Q.shape[1]
    if i >= n:
        return
    if keep_mask[i] == 0 or labels[i] < 0:
        return

    current_label = labels[i]
    best_label = current_label

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0
        for k in range(d):
            diff = int(Q[i, k]) - int(Q[j, k])
            dist2 += diff * diff
            if dist2 > eps2_q:
                break

        if dist2 <= eps2_q:
            lj = labels[j]
            if lj >= 0 and lj < best_label:
                best_label = lj

    if best_label < current_label:
        labels[i] = best_label
        changed[0] = 1


@cuda.jit
def assign_points_to_nearest_core_uint8_kernel(Q, eps2_q, keep_mask, labels):
    i = cuda.grid(1)
    n = Q.shape[0]
    d = Q.shape[1]
    if i >= n:
        return
    if labels[i] >= 0:
        return

    best_label = -1
    best_dist2 = eps2_q
    found = 0

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0
        for k in range(d):
            diff = int(Q[i, k]) - int(Q[j, k])
            dist2 += diff * diff
            if dist2 > eps2_q:
                break

        if dist2 <= eps2_q and (found == 0 or dist2 < best_dist2):
            found = 1
            best_dist2 = dist2
            best_label = labels[j]

    labels[i] = best_label
```

```python
# ============================================================
# 12. Wrapper uint8
# ============================================================

def dbscan_numba_uint8(X, eps, min_samples, keep_ratio=1.0, max_iter=32, seed=SEED):
    if not cuda.is_available():
        raise RuntimeError("CUDA não disponível pelo Numba.")

    start = time.perf_counter()
    Q = quantize_uint8(X)
    n = Q.shape[0]
    blocks = max(1, (n + TPB - 1) // TPB)

    eps_q = float(eps) * 255.0
    eps2_q = np.int32(math.ceil(eps_q * eps_q))

    d_Q = cuda.to_device(Q)
    d_core_flags = cuda.device_array(n, dtype=np.int32)
    d_neighbor_counts = cuda.device_array(n, dtype=np.int32)

    count_core_uint8_kernel[blocks, TPB](d_Q, eps2_q, int(min_samples), d_core_flags, d_neighbor_counts)
    cuda.synchronize()

    core_flags = d_core_flags.copy_to_host()
    keep_mask = make_core_keep_mask(core_flags, keep_ratio=keep_ratio, seed=seed)

    d_keep_mask = cuda.to_device(keep_mask)
    d_labels = cuda.device_array(n, dtype=np.int32)
    init_labels_kernel[blocks, TPB](d_core_flags, d_keep_mask, d_labels)
    cuda.synchronize()

    iterations = 0
    for _ in range(int(max_iter)):
        iterations += 1
        d_changed = cuda.to_device(np.zeros(1, dtype=np.int32))
        propagate_core_labels_uint8_kernel[blocks, TPB](d_Q, eps2_q, d_keep_mask, d_labels, d_changed)
        cuda.synchronize()
        if d_changed.copy_to_host()[0] == 0:
            break

    assign_points_to_nearest_core_uint8_kernel[blocks, TPB](d_Q, eps2_q, d_keep_mask, d_labels)
    cuda.synchronize()

    labels = relabel_consecutive(d_labels.copy_to_host())
    elapsed = time.perf_counter() - start

    n_core = int(core_flags.sum())
    n_kept = int(keep_mask.sum())
    info = {
        "n_core_points": n_core,
        "n_core_kept": n_kept,
        "keep_ratio_real": float(n_kept / max(n_core, 1)),
        "iterations": iterations,
        "keep_ratio_target": float(keep_ratio),
        "quantizacao": "uint8",
        "eps2_quantizado": int(eps2_q),
    }
    return labels, elapsed, info
```

## Versão `uint4 packed`

Os dados são normalizados para `[0, 1]`, convertidos para inteiros de `0` a `15`, e dois atributos de 4 bits são empacotados em um byte. O `eps` é multiplicado por `15`. Esta versão lê menos memória, mas perde mais precisão e paga o custo de desempacotar valores.


```python
# ============================================================
# 13. Quantização uint4 packed e kernels
# ============================================================

def quantize_pack_uint4(X):
    Q4 = np.rint(np.clip(X, 0.0, 1.0) * 15.0).astype(np.uint8)
    n, d = Q4.shape
    packed_d = (d + 1) // 2
    P = np.zeros((n, packed_d), dtype=np.uint8)

    for k in range(d):
        byte_idx = k // 2
        if k % 2 == 0:
            P[:, byte_idx] |= Q4[:, k] & 0x0F
        else:
            P[:, byte_idx] |= (Q4[:, k] & 0x0F) << 4

    return np.ascontiguousarray(P), int(d)


@cuda.jit(device=True)
def get_uint4_value(P, i, k):
    byte_idx = k // 2
    value = P[i, byte_idx]
    if k % 2 == 0:
        return int(value & 0x0F)
    return int((value >> 4) & 0x0F)


@cuda.jit
def count_core_uint4_kernel(P, original_d, eps2_q, min_samples, core_flags, neighbor_counts):
    i = cuda.grid(1)
    n = P.shape[0]
    if i >= n:
        return

    count = 0
    for j in range(n):
        dist2 = 0
        for k in range(original_d):
            a = get_uint4_value(P, i, k)
            b = get_uint4_value(P, j, k)
            diff = a - b
            dist2 += diff * diff
            if dist2 > eps2_q:
                break
        if dist2 <= eps2_q:
            count += 1
            if count >= min_samples:
                break

    neighbor_counts[i] = count
    if count >= min_samples:
        core_flags[i] = 1
    else:
        core_flags[i] = 0


@cuda.jit
def propagate_core_labels_uint4_kernel(P, original_d, eps2_q, keep_mask, labels, changed):
    i = cuda.grid(1)
    n = P.shape[0]
    if i >= n:
        return
    if keep_mask[i] == 0 or labels[i] < 0:
        return

    current_label = labels[i]
    best_label = current_label

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0
        for k in range(original_d):
            a = get_uint4_value(P, i, k)
            b = get_uint4_value(P, j, k)
            diff = a - b
            dist2 += diff * diff
            if dist2 > eps2_q:
                break

        if dist2 <= eps2_q:
            lj = labels[j]
            if lj >= 0 and lj < best_label:
                best_label = lj

    if best_label < current_label:
        labels[i] = best_label
        changed[0] = 1


@cuda.jit
def assign_points_to_nearest_core_uint4_kernel(P, original_d, eps2_q, keep_mask, labels):
    i = cuda.grid(1)
    n = P.shape[0]
    if i >= n:
        return
    if labels[i] >= 0:
        return

    best_label = -1
    best_dist2 = eps2_q
    found = 0

    for j in range(n):
        if keep_mask[j] == 0 or labels[j] < 0:
            continue

        dist2 = 0
        for k in range(original_d):
            a = get_uint4_value(P, i, k)
            b = get_uint4_value(P, j, k)
            diff = a - b
            dist2 += diff * diff
            if dist2 > eps2_q:
                break

        if dist2 <= eps2_q and (found == 0 or dist2 < best_dist2):
            found = 1
            best_dist2 = dist2
            best_label = labels[j]

    labels[i] = best_label
```

```python
# ============================================================
# 14. Wrapper uint4 packed
# ============================================================

def dbscan_numba_uint4(X, eps, min_samples, keep_ratio=1.0, max_iter=32, seed=SEED):
    if not cuda.is_available():
        raise RuntimeError("CUDA não disponível pelo Numba.")

    start = time.perf_counter()
    P, original_d = quantize_pack_uint4(X)
    n = P.shape[0]
    blocks = max(1, (n + TPB - 1) // TPB)

    eps_q = float(eps) * 15.0
    eps2_q = np.int32(math.ceil(eps_q * eps_q))

    d_P = cuda.to_device(P)
    d_core_flags = cuda.device_array(n, dtype=np.int32)
    d_neighbor_counts = cuda.device_array(n, dtype=np.int32)

    count_core_uint4_kernel[blocks, TPB](d_P, original_d, eps2_q, int(min_samples), d_core_flags, d_neighbor_counts)
    cuda.synchronize()

    core_flags = d_core_flags.copy_to_host()
    keep_mask = make_core_keep_mask(core_flags, keep_ratio=keep_ratio, seed=seed)

    d_keep_mask = cuda.to_device(keep_mask)
    d_labels = cuda.device_array(n, dtype=np.int32)
    init_labels_kernel[blocks, TPB](d_core_flags, d_keep_mask, d_labels)
    cuda.synchronize()

    iterations = 0
    for _ in range(int(max_iter)):
        iterations += 1
        d_changed = cuda.to_device(np.zeros(1, dtype=np.int32))
        propagate_core_labels_uint4_kernel[blocks, TPB](d_P, original_d, eps2_q, d_keep_mask, d_labels, d_changed)
        cuda.synchronize()
        if d_changed.copy_to_host()[0] == 0:
            break

    assign_points_to_nearest_core_uint4_kernel[blocks, TPB](d_P, original_d, eps2_q, d_keep_mask, d_labels)
    cuda.synchronize()

    labels = relabel_consecutive(d_labels.copy_to_host())
    elapsed = time.perf_counter() - start

    n_core = int(core_flags.sum())
    n_kept = int(keep_mask.sum())
    info = {
        "n_core_points": n_core,
        "n_core_kept": n_kept,
        "keep_ratio_real": float(n_kept / max(n_core, 1)),
        "iterations": iterations,
        "keep_ratio_target": float(keep_ratio),
        "quantizacao": "uint4_packed",
        "eps2_quantizado": int(eps2_q),
        "original_d": int(original_d),
        "packed_d": int(P.shape[1]),
    }
    return labels, elapsed, info
```

Esta celula roda, no dataset ativo:

Observacao importante: se `RUN_MODE = "quick"`, os tempos servem para depuracao e comparacao didatica. Para usar numeros no relatorio/apresentacao, prefira `RUN_MODE = "presentation"`, porque datasets pequenos subutilizam a GPU e podem exagerar o speedup contra cuML por causa do overhead de inicializacao do cuML.


```python
# ============================================================
# 15. Rodar todos os experimentos
# ============================================================

MAX_PROPAGATION_ITERS = 32
EXPERIMENT_SEED = 42


def warmup_numba_kernels():
    """Compila kernels antes da medição principal para reduzir viés de JIT."""
    if not cuda.is_available():
        print("Warm-up ignorado: CUDA indisponível.")
        return

    rng = np.random.default_rng(EXPERIMENT_SEED)
    Xw = np.ascontiguousarray(rng.random((64, min(4, X.shape[1])), dtype=np.float32))
    _ = dbscan_numba_float32(Xw, 0.35, 4, keep_ratio=1.0, max_iter=2, seed=EXPERIMENT_SEED)
    _ = dbscan_numba_uint8(Xw, 0.35, 4, keep_ratio=1.0, max_iter=2, seed=EXPERIMENT_SEED)
    _ = dbscan_numba_uint4(Xw, 0.35, 4, keep_ratio=1.0, max_iter=2, seed=EXPERIMENT_SEED)
    print("Warm-up Numba CUDA concluído.")


results = []
all_labels = {}

print(f"Dataset: {DATASET_NAME}")
print(f"RUN_MODE: {RUN_MODE}")
if RUN_MODE == "quick":
    print("Timing note: quick mode is didactic; for report numbers use RUN_MODE = 'presentation'.")
print(f"shape={X.shape}, eps={EPS:.6f}, min_samples={MIN_SAMPLES}")

labels_baseline, time_baseline, backend_baseline, info_baseline = run_baseline_dbscan(X, EPS, MIN_SAMPLES)
baseline_key = f"baseline_{backend_baseline}"
all_labels[baseline_key] = labels_baseline
results.append(build_result_row(baseline_key, labels_baseline, time_baseline, labels_baseline, info_baseline))
print(f"{baseline_key}: {time_baseline:.4f}s")

warmup_numba_kernels()

gpu_experiments = [
    ("numba_float32_keep_100", dbscan_numba_float32, {"keep_ratio": 1.00}),
    ("drop_core_keep_75", dbscan_numba_float32, {"keep_ratio": 0.75}),
    ("drop_core_keep_50", dbscan_numba_float32, {"keep_ratio": 0.50}),
    ("drop_core_keep_25", dbscan_numba_float32, {"keep_ratio": 0.25}),
    ("numba_uint8_keep_100", dbscan_numba_uint8, {"keep_ratio": 1.00}),
    ("numba_uint4_packed_keep_100", dbscan_numba_uint4, {"keep_ratio": 1.00}),
]

for version, func, kwargs in gpu_experiments:
    try:
        labels, elapsed, info = func(
            X,
            EPS,
            MIN_SAMPLES,
            max_iter=MAX_PROPAGATION_ITERS,
            seed=EXPERIMENT_SEED,
            **kwargs,
        )
        all_labels[version] = labels
        results.append(build_result_row(version, labels, elapsed, labels_baseline, info))
        print(f"{version}: {elapsed:.4f}s")
    except Exception as exc:
        print(f"{version} não executado: {repr(exc)}")
```

## Discussao curta dos resultados esperados

Em datasets densos, espera-se observar muitos core points. Isso tende a aumentar o custo de conexao entre cores, pois muitos pontos tem vizinhancas sobrepostas. A versao com drop tenta reduzir esse custo mantendo apenas uma fracao dos cores e atribuindo os demais pontos ao core mantido mais proximo dentro de `eps`.

O trade-off esperado e direto: quanto menor o `keep_ratio`, maior a chance de reduzir tempo, mas maior o risco de fragmentar clusters, aumentar ruido ou reduzir ARI/NMI contra o baseline. As versoes `uint8` e `uint4 packed` reduzem leitura de memoria por atributo, o que pode ajudar mais em datasets de maior dimensao, como `blobs_32d`; em troca, elas alteram as distancias por quantizacao.

Atencao na interpretacao dos tempos: em `RUN_MODE = "quick"`, o dataset e pequeno e a GPU pode ficar subutilizada. Alem disso, o tempo do cuML pode incluir overhead de inicializacao. Portanto, speedups muito altos nesse modo devem ser tratados como evidencia exploratoria, nao como conclusao final. Para relatorio, rode tambem com `RUN_MODE = "presentation"` e compare ARI/NMI, ruido e numero de clusters.

Limitacoes: esta e uma implementacao didatica, com vizinhanca por forca bruta `O(n^2)`, sem indice espacial, sem union-find otimizado e com propagacao iterativa limitada por `MAX_PROPAGATION_ITERS`. Ela serve para estudar as ideias do trabalho, nao para substituir uma implementacao industrial como cuML. Multi-EPS fica como etapa futura.
